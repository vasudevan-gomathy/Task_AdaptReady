2) Given an unsorted array of integers, find the length of the longest continuous increasing subsequence (subarray). 

Example 1:
Input: [1,3,5,4,7]
Output: 3 

Example 2:
Input: [2,2,2,2,2]
Output: 1

```python
## INPUT:

nums = [1,3,5,4,7]

## SOLUTION:

"""
Problem Statement: Finding length of longest subsequence.

PROBLEM(1): Getting length of longest subsequence by comparing previous subsequence length.
PROBLEM(2): If current item is less than previous item, subsequence should startover from current item.
"""

# Code:

l = 0 # startover subsequence index
longest_sub_len = 0 # our desirable output
prev_item = 0 # previous item

for r in range(len(nums)):
  if nums[r] > prev_item:
    longest_sub_len = max(longest_sub_len, r - l + 1) # SOLUTION(1)
  else:
    l = r # SOLUTION(2)
  
  prev_item = nums[r]

## OUTPUT:
print(longest_sub_len)
```

3. Given a list of non negative integers, arrange them such that they form the largest number.
 
 Example 1:
 Input: [10,2]
 Output: 210

 Example 2:
 Input: [3,30,34,5,9]
 Output: 9534330

```python
## INPUT:

nums = [3,30,34,5,9]

## SOLUTION:

"""
Problem Statement: Finding the largest number formed from the given non-negative intergers.

PROBLEM(1): Finding the best number by comparing two integers
For Example:
Between 9 and 5 ==> Comparing is easy as 9 is greater than 5
But for multi-digit integers like 5 and 34 ==> Comparing should be done as "534" or "345"

Used "cmp_to_key" for comparing the concatenated element

PROBLEM(2): When all the given number is 0, result should be "0" rather than "00000"
"""

# Code:

import functools

nums = [3,30,34,5,9]

for i,n in enumerate(nums): # Converting int into str for concatenation used for comparison.
  nums[i] = str(n)

def compare(x1, x2):
  if x1 + x2 > x2 + x1:
    return -1
  else:
    return 1

nums = sorted(nums, key=functools.cmp_to_key(compare)) # SOLUTION(1)

## OUTPUT:

print(str(int("".join(nums)))) # SOLUTION(2)
```

4. Store all the "servlet-name", and "servlet-class" to a csv file from the attached sample_json.json file using Python.

```python
# Import necessary packages
import json
import csv

# Load the JSON data
with open('sample_json.json', 'r') as f:
    data = json.load(f)

# Extract "servlet-name" and "servlet-class" from each entry
servlet_data = [(servlet['servlet-name'], servlet['servlet-class']) for servlet in data['web-app']['servlet']]

# Write the data to a CSV file
with open('servlet_data.csv', 'w', newline='') as csvfile:
    fieldnames = ['servlet-name', 'servlet-class']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

    writer.writeheader()
    for servlet_name, servlet_class in servlet_data:
        writer.writerow({'servlet-name': servlet_name, 'servlet-class': servlet_class})

print("CSV file generated successfully!")
```
